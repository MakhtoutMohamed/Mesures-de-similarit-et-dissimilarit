# Atelier 2 — Mesures de similarité et dissimilarité (Mall_Customers)

```python

import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances, cosine_similarity
from scipy.stats import spearmanr

df = pd.read_csv('Mall_Customers.csv')
print('Head:'); display(df.head())
print('Tail:'); display(df.tail())

# k$ -> $
df['Annual Income ($)'] = df['Annual Income (k$)'] * 1000
features = ['Age','Annual Income ($)','Spending Score (1-100)']
data = df[features].copy()

# distances (raw)
dist_euc = euclidean_distances(data)
dist_man = manhattan_distances(data)

# normalization
scaler = MinMaxScaler(); data_n = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
dist_euc_n = euclidean_distances(data_n); dist_man_n = manhattan_distances(data_n)

# similarities
cos_sim = cosine_similarity(data_n)
corr_feat = data_n.corr(method='pearson')

# compare Euclidean vs (1 - Cosine)
def upper_tri(v):
    i = np.triu_indices_from(v, k=1); return v[i]
from scipy.stats import spearmanr
rho, p = spearmanr(upper_tri(dist_euc_n), upper_tri(1 - cos_sim))
print(f"Spearman corr (Euclidean vs 1-Cosine): rho={rho:.3f}, p={p:.3e}")

# simple heatmaps (matplotlib only; one figure per plot)
def show_heatmap(M, title, xl='Index', yl='Index'):
    plt.figure(figsize=(6,5))
    plt.imshow(M, aspect='auto')
    plt.title(title); plt.colorbar(); plt.xlabel(xl); plt.ylabel(yl); plt.show()

show_heatmap(dist_euc_n, "Distance euclidienne (normalisée)")
show_heatmap(cos_sim, "Similarité cosinus (observations)")
show_heatmap(corr_feat.values, "Corrélation (entre variables)", "Feature", "Feature")
```
